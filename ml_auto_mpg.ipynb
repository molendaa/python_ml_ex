```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
auto = pd.read_csv(r"auto-mpg.csv")

auto.shape
auto.head()
```

```python
auto.drop(auto[auto['horsepower'] == '?'].index, inplace = True)

X = auto.iloc[:,1:-1]
#X = X.drop(columns = ['horsepower'])
X.head()
#x_error = X.loc[X['horsepower'] == '?']
#x_error
y = auto.iloc[:,0]
y.head()

lr = LinearRegression()
lr.fit(X.to_numpy(),y)
print(lr.score(X.to_numpy(),y))


# my_car1 = [4, 160, 190, 12, 90, 1]
# my_car2 = [4, 200, 260, 15, 83, 1]

my_car1 = [4, 160, 190, 3000, 12, 90, 1]
my_car2 = [4, 200, 260, 2500, 15, 83, 1]
 
cars = [my_car1, my_car2]

mpg_predict = lr.predict(cars)
print(mpg_predict)
```
